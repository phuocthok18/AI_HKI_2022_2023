```python
import numpy as np
import pandas as pd
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split

data = pd.read_csv('/content/drive/MyDrive/data_ai/crabs.csv')
data.head(1000)
```

```python
# xoa o trong speies
data = data[data['speies'] != 'O']
# thay thế gioi tinh M=0, F=1
data = data.replace('M',1)
data = data.replace('F',0)

x = data.drop(['speies', 'sex', 'index'], axis = 1)
y = data.drop(['speies', 'index', 'fontallip', 'rearwidth', 'length', 'width', 'depth'], axis = 1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
model = Perceptron()
model.fit(x_train, y_train)
score_train = model.score(x_train, y_train)
print('do chinh xac',score_train)
y_het = model.predict(x_test)
cal = pd .DataFrame(np.c_[y_test, y_het],columns = ['tested y','predicted y'])
print(cal)
data.head(1000)
```

```python
# hoa lan iris
import numpy as np
from sklearn.linear_model import Perceptron
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
# from sklearn.metrics import accuracy_score
import pandas as pd

iris = load_iris()
x = iris.data
y = iris.target
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
model = Perceptron()
model.fit(x_train, y_train)
score_train = model.score(x_train, y_train)
print(score_train)
y_het = model.predict(x_test)
cal = pd .DataFrame(np.c_[y_test, y_het],columns = ['tested y','predicted y'])
print(cal)
```

```python
import numpy as np
from sklearn.linear_model import Perceptron
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
iris = load_iris()
for n, i in enumerate(iris.target):
  if i==2:
    iris.target[n]=1
x=iris.data
y=iris.target
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
model = Perceptron()
model.fit(x_train, y_train)
score_train = model.score(x_test, y_test)
print('do chinh xac test:',score_train)
x_test = [[4,3,2,0.5],[5,7,3,0.6]]
y_test = model.predict(x_test)
print(y_test)
```
